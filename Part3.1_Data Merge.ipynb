```python
import pandas as pd
```

## 3.2 Seller merge

```python
seller_features = pd.read_csv("data_processed/seller_features.csv")
seller_features.head()
```

```python
seller_overall = pd.read_csv('data_processed/seller_overall.csv')
seller_overall.head()
```

```python
seller_repeat = pd.read_csv('data_processed/repeat_merchant_result.csv')
seller_repeat.head()
```

```python
seller_features.shape
```

```python
seller_overall.shape
```

```python
seller_repeat.shape
```

```python
seller_FINAL = pd.merge(seller_features, seller_overall, on='seller_id')
seller_FINAL = pd.merge(seller_FINAL, seller_repeat, how='left',on='seller_id').fillna(0)
```

```python
seller_FINAL.shape
```

### column cleaning

```python
seller_FINAL.columns
```

```python
seller_FINAL = seller_FINAL.drop(['Unnamed: 0_x','Unnamed: 0_y'], axis=1)
```

```python
#data output
seller_FINAL.to_csv('Final_data/seller_FINAL.csv')
```

## 3.3 <seller, user> merge

```python
su_overall = pd.read_csv("data_processed/su_overall.csv")
su_overall.head()
```

```python
su_double11 = pd.read_csv("data_processed/Double11_su.csv")
su_double11.head()
```

```python
su_pd = pd.read_csv("data_processed/su_pd.csv")
su_pd.head()
```

```python
su_repeat = pd.read_csv("data_processed/su_repeat.csv")
su_repeat.head()
```

```python
su_FINAL = pd.merge(su_overall, su_double11, how ='left', on='su_id')
```

```python
su_FINAL = pd.merge(su_FINAL, su_pd, how ='left', on='su_id')
```

```python
su_FINAL = pd.merge(su_FINAL, su_repeat, how ='left', on='su_id')
```

### column cleaning

```python
su_FINAL.columns
```

```python
su_FINAL = su_FINAL.drop(['Unnamed: 0_x','Unnamed: 0_y', 'Unnamed: 0'], axis=1)
```

```python
#data output
su_FINAL.to_csv('Final_data/su_FINAL.csv')
```
